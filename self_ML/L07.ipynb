```python
#MNIST 패션 데이터셋 불러오기

from tensorflow import keras
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
```

```python
#이미지 픽셀값 0~255 -> 0~1 사이로 변환
#28*28 크기의 2차원 배열을 784 크기의 1차원 배열로 펼침
#훈련, 검증 세트로 나누기

from sklearn.model_selection import train_test_split
train_scaled = train_input / 255.0
train_scaled = train_scaled.reshape(-1, 28*28)
train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, train_target, test_size=0.2, random_state=42)
```

은닉층에 비선형 활성화함수 사용

why?)

선형적 관계가 되면 사실상 층을 쌓는 의미가 없어짐(대입하면 되니까)



```python
#2개의 층 쌓기

#은닉층, 100개의 뉴런을 가진 밀집층
dense1 = keras.layers.Dense(100, activation='sigmoid', input_shape=(784, ))

#10개의 클래스를 분류하므로, 10개의 뉴런
dense2 = keras.layers.Dense(10, activation='softmax')
```

```python
#심층신경망 만들기

#층 순서대로 나열
model=keras.Sequential([dense1, dense2])
```

```python
#층에 대한 정보 확인

model.summary()
```

```python
#Sequential 클래스에 층을 추가하는 다른 방법

model = keras.Sequential([keras.layers.Dense(100, activation='sigmoid', input_shape=(784,), name='hidden'), keras.layers.Dense(10, activation='softmax', name='output')], name='패션 MNIST 모델')
```

```python
#이름 지정 후, 층에 대한 정보 확인

model.summary()
```

```python
#add를 사용하여 층 추가하기

model=keras.Sequential()

model.add(keras.layers.Dense(100, activation='sigmoid', input_shape=(784,)))
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
#모델 정보 확인

model.summary()
```

```python
#5번의 에포크로 모델 훈련

model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy')
model.fit(train_scaled, train_target, epochs=5)
```

양 끝이 누워있는 형태라서 층이 깊어질 수록 '가중치 소실' 문제 발생 가능

07-3 신경망 모델 훈련

머신러닝 : 매개변수를 조정하고, 훈련하는 과정 반복 (모델의 구조가 고정되어 있음)

딥러닝 : 모델의 구조를 직접 구성

```python
#History 클래스 객체 : 훈련과정에서 계산한 여러 정보들이 저장되어 있음

#패션 MNIST 데이터 훈련, 검증 세트로 분할
from tensorflow import keras
from sklearn.model_selection import train_test_split
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

train_scaled = train_input / 255.0

train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
#모델 생성(함수 이용)

def model_fn(a_layer=None):
    model = keras.Sequential()
    model.add(keras.layers.Flatten(input_shape=(28,28)))
    model.add(keras.layers.Dense(100, activation='relu'))
    
    if a_layer:
        model.add(a_layer)
        
    model.add(keras.layers.Dense(10, activation='softmax'))
    return model
```

```python
#모델의 구조 확인
#이전에 만든 것과 동일한 모델

model = model_fn()
model.summary()
```

```python
#fit() 메서드의 결과를 history 변수에 저장

model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy')
history = model.fit(train_scaled, train_target, epochs=5, verbose=0)
```

```python
#훈련 측정값이 담겨 있는 history 딕셔너리 확인

print(history.history.keys())
```

```python
#history에 포함된 손실, 정확도 -> 에포크마다 계산한 값이 순서대로 나열된 리스트
#그래프로 시각화

import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

```python
#정확도 시각화

plt.plot(history.history['accuracy'])
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

에포크가 진행될 수록 손실은 작아지고, 정확도는 높아지는 것을 확인할 수 있음

-> 모델이 잘 훈련되고 있다는 뜻

```python
#에포크 횟수를 늘려서 모델 훈련

model=model_fn()
model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy')
history = model.fit(train_scaled, train_target, epochs=20, verbose=0)
plt.plot(history.history['loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

에포크 수를 늘렸더니 손실이 더욱 줄어듦

검증손실)

에포크와 과대/과소적합의 관계성

-> 검증세트에 대한 점수도 필요

```python
#에포크마다 검증 손실 계산
#매개변수에 검증에 사용할 입력,타깃값을 튜플로 만들어 전달

model = model_fn()
model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy')
history = model.fit(train_scaled, train_target, epochs=20, verbose=0, validation_data=(val_scaled, val_target))
```

```python
#history.history 딕셔너리의 키 확인

print(history.history.keys())
```

```python
#훈련, 검증 손실을 한그래프에 그려서 비교

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
```

초기에는 전체적으로 검증손실이 감소하다가, 9번째 에포크부터 검증 손실이 크게 증가함

훈련손실을 꾸준히 감소함

-> 과대적합된 모델

*검증손실이 상승하는 지점을 최대한 뒤로 늦추기 -> 검증세트에 대한 정확도 증가

```python
#과대적합을 막기 위해 규제 사용
#RMSprop 대신에 Adam 옵티마이저 사용
#Adam은 적응적 학습률을 사용하므로, 에포크가 진행됨에 따라 학습률 조정 가능

model = model_fn()
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
history = model.fit(train_scaled, train_target, epochs=20, verbose=0, validation_data=(val_scaled, val_target))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
```

adam 옵티마이저를 사용하니 과대적합이 훨씬 줄어듦

*16번째 에포크까지는 감소하는 추세가 이어짐

- 학습률을 조정해서 더 나은 손실곡선 얻기

```python
#adam 옵티마이저 학습률 조정
#학습률 : 기본 0.001 -> 조정 0.0005

from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# 모델 생성
model = model_fn()

# Adam 옵티마이저의 학습률 조정 (예: 0.0005)
optimizer = Adam(learning_rate=0.0005)

# 모델 컴파일
model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 학습 진행
history = model.fit(train_scaled, train_target, epochs=20, verbose=0, validation_data=(val_scaled, val_target))

# 손실 그래프 출력
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
```

학습률을 0.0005로 조정했더니 검증손실 그래프가 확실히 개선됨

```python
#학습률 스케줄링 (동적으로 학습률 조정)
#학습이 진행되면서 학습률을 점점 낮추는 방식

from tensorflow.keras.optimizers.schedules import ExponentialDecay

# 학습률을 점점 감소시키는 스케줄 설정
lr_schedule = ExponentialDecay(initial_learning_rate=0.001, decay_steps=1000, decay_rate=0.9)

# Adam 옵티마이저 적용
optimizer = Adam(learning_rate=lr_schedule)

# 모델 컴파일
model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])


# 손실 그래프 출력
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
```

단계적으로 학습률을 줄이는 방식도 확실히 개선된 검증손실 그래프를 보여줌

```python
#ReduceLROnPlateau (손실이 줄어들지 않으면 학습률 감소)

from tensorflow.keras.callbacks import ReduceLROnPlateau

# 학습률 감소 콜백 설정 (val_loss가 3번 연속 감소하지 않으면 0.5배 감소)
lr_reducer = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)

# 모델 학습
history = model.fit(train_scaled, train_target, epochs=20, verbose=0, validation_data=(val_scaled, val_target), callbacks=[lr_reducer])

# 손실 그래프 출력
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
```

-> 즉 모델을 2번씩 훈련함

```python
#콜백
#훈련 과정 중간에 어떤 작업을 수행할 수 있게 하는 객체
#에포크 마다 모델을 저장 -> 가장 낮은 검증 점수를 만드는 모델을 저장할 수 있음

#모델 생성
model = model_fn(keras.layers.Dropout(0.3))

#모델 컴파일(모델 학습 준비)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')

#모델 학습을 멈출 조건 설정(에포크가 남았더라도 종료)
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-model.h5', save_best_only=True)

#모델 훈련
model.fit(train_scaled, train_target, epochs=20, verbose=0, validation_data=(val_scaled, val_target), callbacks=[checkpoint_cb])
```

```python
#최상의 점수를 낸 모델이 저장되었으니, 이 모델로 예측 수행

model = keras.models.load_model('best-model.h5')
model.evaluate(val_scaled, val_target)
```

ModelCheckpoint 콜백이 가장 낮은 검증 점수의 모델을 자동으로 저장해줌


조기종료)

검증 점수가 상승하기 시작하면 그 이후에는 과대적합이 더 커지므로, 훈련을 계속할 필요가 없음

-> 이때 훈련을 종료하면 자원, 시간 절약 가능

: 에포크 수를 제한하는 역할(규제의 일종)

조기종료, 콜백 사용

-> 가장 낮은 검증 손실의 모델을 파일에 저장하고, 검증 손실이 다시 상승할 때 훈련 중지 가능

-> 또, 훈련을 중지한 다음 현재 모델의 파라미터를 최상의 파라미터로 되돌림

```python
model = model_fn(keras.layers.Dropout(0.3))

#모델 컴파일(모델 학습 준비)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')

#콜백 1
#최적의 모델 저장
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-model.h5', save_best_only=True)

#콜백 2
#patience 매개변수 : 검증점수가 향상 되지 않더라도 참을 에포크 횟수 지정
#restore_best_weights 매개변수 : 가장 낮은 검증 손실을 낸 모델 파라미터로 되돌림
early_stopping_cb = keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)

#모델 훈련 및 정보 저장
#2개의 콜백을 리스트로 전달함
history = model.fit(train_scaled, train_target, epochs=20, verbose=0, validation_data=(val_scaled, val_target), callbacks=[checkpoint_cb, early_stopping_cb])
```

```python
#몇번째 에포크에서 훈련이 중지되었는지 확인

print(early_stopping_cb.stopped_epoch)
```

*에포크는 0 부터 시작 -> 11번째 에포크에서 훈련이 중지되었다는 뜻

patience=2이므로, 최적의 모델은 9번째 에포크(에포크 8)

```python
#훈련, 검증 손실 확인

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
```

9번째 에포크(에포크 8)에서 가장 낮은 검증 손실을 기록

11번째 에포크(에포크 10)에서 훈련 중지(조기종료)

-> 조기종료 기법을 사용하면 에포크 수를 크기 지정해도 됨

- EarlyStopping 조기종료  -> 자원, 시간 절약

- ModelCheckpoint 콜백과 함께 사용 -> 최상의 모델을 자동으로 저장

```python
#조기종료로 얻은 모델을 사용해 검증세트에 대한 성능 확인

model.evaluate(val_scaled, val_target)
```
